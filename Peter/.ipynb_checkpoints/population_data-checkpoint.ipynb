```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import numpy as np
import json
import csv
```

```python
year_2014 = pd.read_csv('Y2014_pop_data_with_overlays.csv',header=1)
year_2015 = pd.read_csv('Y2015_pop_data_with_overlays.csv',header=1)
year_2016 = pd.read_csv('Y2016_pop_data_with_overlays.csv',header=1)
year_2017 = pd.read_csv('Y2017_pop_data_with_overlays.csv',header=1)
year_2018 = pd.read_csv('Y2018_pop_data_with_overlays.csv',header=1)
```

```python
year_2014 = year_2014.dropna()
year_2015 = year_2015.dropna()
year_2016 = year_2016.dropna()
year_2017 = year_2017.dropna()
year_2018 = year_2018.dropna()
```

```python
year_2018
```

```python
year_2014["Other"] = year_2014['Estimate!!Total!!Native Hawaiian and Other Pacific Islander alone']+year_2014['Estimate!!Total!!Some other race alone']+year_2014['Estimate!!Total!!Two or more races']
year_2015["Other"] = year_2015['Estimate!!Total!!Native Hawaiian and Other Pacific Islander alone']+year_2015['Estimate!!Total!!Some other race alone']+year_2015['Estimate!!Total!!Two or more races']
year_2016["Other"] = year_2016['Estimate!!Total!!Native Hawaiian and Other Pacific Islander alone']+year_2016['Estimate!!Total!!Some other race alone']+year_2016['Estimate!!Total!!Two or more races']
year_2017["Other"] = year_2017['Estimate!!Total!!Native Hawaiian and Other Pacific Islander alone']+year_2017['Estimate!!Total!!Some other race alone']+year_2017['Estimate!!Total!!Two or more races']
year_2018["Other"] = year_2018['Estimate!!Total!!Native Hawaiian and Other Pacific Islander alone']+year_2018['Estimate!!Total!!Some other race alone']+year_2018['Estimate!!Total!!Two or more races']
```

```python
year_2014_reduced_columns = year_2014[
    ['id',
     'Geographic Area Name',
     'Estimate!!Total!!White alone',
     'Estimate!!Total!!Black or African American alone',
     'Estimate!!Total!!American Indian and Alaska Native alone',
     'Estimate!!Total!!Asian alone',
     'Other'
    ]]

year_2015_reduced_columns = year_2015[
    ['id',
     'Geographic Area Name',
     'Estimate!!Total!!White alone',
     'Estimate!!Total!!Black or African American alone',
     'Estimate!!Total!!American Indian and Alaska Native alone',
     'Estimate!!Total!!Asian alone',
     'Other'
    ]]

year_2016_reduced_columns = year_2016[
    ['id',
     'Geographic Area Name',
     'Estimate!!Total!!White alone',
     'Estimate!!Total!!Black or African American alone',
     'Estimate!!Total!!American Indian and Alaska Native alone',
     'Estimate!!Total!!Asian alone',
     'Other'
    ]]

year_2017_reduced_columns = year_2017[
    ['id',
     'Geographic Area Name',
     'Estimate!!Total!!White alone',
     'Estimate!!Total!!Black or African American alone',
     'Estimate!!Total!!American Indian and Alaska Native alone',
     'Estimate!!Total!!Asian alone',
     'Other'
    ]]

year_2018_reduced_columns = year_2018[
    ['id',
     'Geographic Area Name',
     'Estimate!!Total!!White alone',
     'Estimate!!Total!!Black or African American alone',
     'Estimate!!Total!!American Indian and Alaska Native alone',
     'Estimate!!Total!!Asian alone',
     'Other'
    ]]
```

```python
pop_2014_df = year_2014_reduced_columns.rename(columns={
    "id": "id",
    "Geographic Area Name": "county",
    'Estimate!!Total!!White alone': "white_population",
    'Estimate!!Total!!Black or African American alone': "african_american_pop",
    'Estimate!!Total!!American Indian and Alaska Native alone': "american_indian_alaska_native_pop",
    'Estimate!!Total!!Asian alone': "asian_pop"
    })

pop_2015_df = year_2015_reduced_columns.rename(columns={
    "id": "id",
    "Geographic Area Name": "county",
    'Estimate!!Total!!White alone': "white_population",
    'Estimate!!Total!!Black or African American alone': "african_american_pop",
    'Estimate!!Total!!American Indian and Alaska Native alone': "american_indian_alaska_native_pop",
    'Estimate!!Total!!Asian alone': "asian_pop"
    })

pop_2016_df = year_2016_reduced_columns.rename(columns={
    "id": "id",
    "Geographic Area Name": "county",
    'Estimate!!Total!!White alone': "white_population",
    'Estimate!!Total!!Black or African American alone': "african_american_pop",
    'Estimate!!Total!!American Indian and Alaska Native alone': "american_indian_alaska_native_pop",
    'Estimate!!Total!!Asian alone': "asian_pop"
    })

pop_2017_df = year_2017_reduced_columns.rename(columns={
    "id": "id",
    "Geographic Area Name": "county",
    'Estimate!!Total!!White alone': "white_population",
    'Estimate!!Total!!Black or African American alone': "african_american_pop",
    'Estimate!!Total!!American Indian and Alaska Native alone': "american_indian_alaska_native_pop",
    'Estimate!!Total!!Asian alone': "asian_pop"
    })

pop_2018_df = year_2018_reduced_columns.rename(columns={
    "id": "id",
    "Geographic Area Name": "county",
    'Estimate!!Total!!White alone': "white_population",
    'Estimate!!Total!!Black or African American alone': "african_american_pop",
    'Estimate!!Total!!American Indian and Alaska Native alone': "american_indian_alaska_native_pop",
    'Estimate!!Total!!Asian alone': "asian_pop"
    })

```

```python
pop_2014_df["Total_Population"] = pop_2014_df['white_population']+pop_2014_df['african_american_pop']+pop_2014_df['american_indian_alaska_native_pop']+pop_2014_df['asian_pop']+pop_2014_df['Other']
pop_2015_df["Total_Population"] = pop_2015_df['white_population']+pop_2015_df['african_american_pop']+pop_2015_df['american_indian_alaska_native_pop']+pop_2015_df['asian_pop']+pop_2015_df['Other']
pop_2016_df["Total_Population"] = pop_2016_df['white_population']+pop_2016_df['african_american_pop']+pop_2016_df['american_indian_alaska_native_pop']+pop_2016_df['asian_pop']+pop_2016_df['Other']
pop_2017_df["Total_Population"] = pop_2017_df['white_population']+pop_2017_df['african_american_pop']+pop_2017_df['american_indian_alaska_native_pop']+pop_2017_df['asian_pop']+pop_2017_df['Other']
pop_2018_df["Total_Population"] = pop_2018_df['white_population']+pop_2018_df['african_american_pop']+pop_2018_df['american_indian_alaska_native_pop']+pop_2018_df['asian_pop']+pop_2018_df['Other']
```

```python
pop_2014_df[['County','State']] = pop_2014_df.county.str.split(",",expand=True,)
pop_2014_df['County'] = pop_2014_df['County'].str.replace(" County", '')
pop_2014_df = pop_2014_df.drop(columns=["county"])

pop_2015_df[['County','State']] = pop_2015_df.county.str.split(",",expand=True,)
pop_2015_df['County'] = pop_2015_df['County'].str.replace(" County", '')
pop_2015_df = pop_2015_df.drop(columns=["county"])

pop_2016_df[['County','State']] = pop_2016_df.county.str.split(",",expand=True,)
pop_2016_df['County'] = pop_2016_df['County'].str.replace(" County", '')
pop_2016_df = pop_2016_df.drop(columns=["county"])

pop_2017_df[['County','State']] = pop_2017_df.county.str.split(",",expand=True,)
pop_2017_df['County'] = pop_2017_df['County'].str.replace(" County", '')
pop_2017_df = pop_2017_df.drop(columns=["county"])

pop_2018_df[['County','State']] = pop_2018_df.county.str.split(",",expand=True,)
pop_2018_df['County'] = pop_2018_df['County'].str.replace(" County", '')
pop_2018_df = pop_2018_df.drop(columns=["county"])
```

```python
geo_loc_counties = pd.read_csv('Geo_Loc_Counties.csv')
geo_loc_counties
```

```python
pop_2014_df = pd.merge(pop_2014_df, geo_loc_counties, on="County")
pop_2015_df = pd.merge(pop_2015_df, geo_loc_counties, on="County")
pop_2016_df = pd.merge(pop_2016_df, geo_loc_counties, on="County")
pop_2017_df = pd.merge(pop_2017_df, geo_loc_counties, on="County")
pop_2018_df = pd.merge(pop_2018_df, geo_loc_counties, on="County")
```

```python
pop_2014_df = pd.merge(pop_2014_df, nate_csv, on= ["County"])
pop_2015_df = pd.merge(pop_2015_df, nate_csv, on= ["County"])
pop_2016_df = pd.merge(pop_2016_df, nate_csv, on= ["County"])
pop_2017_df = pd.merge(pop_2017_df, nate_csv, on= ["County"])
pop_2018_df = pd.merge(pop_2018_df, nate_csv, on= ["County"])
```
